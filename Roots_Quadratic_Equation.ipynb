```python
import cmath

#enter the co-effients:

a=eval(input("the co-efficient of sqr of x: "))
b=eval(input("co-effiecient of x: "))
c=eval(input("constant: "))

if(type(a)==int and type(b)==int and type(c)==int):
    #calculating the roots
    d = (b*b)-(4*a*c)

    root1 = ((-b)-cmath.sqrt(d))/(2*a)
    root2 = ((-b)+cmath.sqrt(d))/(2*a)

    print("first root: ",root1)
    print("second root: ",root2)
else:
    print("enter a valid quadratic equation")
```
